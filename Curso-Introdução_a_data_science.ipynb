Variaveis sao Qualitativas(categoricas) ou Quantitativas(numericas)

**Variaveis Qualitativas**

representam características que não podem ser medidas numericamente, mas podem ser categorizadas, ou seja, representam uma classificação. Estas variáveis podem ser **nominais** ou **ordinais**.

**Nominal**

Uma variável qualitativa nominal é aquela que classifica os elementos em classes ou categorias em relação ao tipo do dado representado. Ela não possui uma relação de grandeza ou ordem dos dados

Exemplos: Nome de produtos, departamentos, métodos de pagamento ou variáveis geográficas, como, cidade, estado e região.

**Ordinal**

Uma variável qualitativa ordinal também é aquela que classifica os elementos em classes ou categorias, porém existe uma ordem natural ou hierárquica entre as categorias.

Exemplos: grau de escolaridade (ensino infantil, ensino fundamental, ensino médio), nível de satisfação do cliente (insatisfeito, neutro e satisfeito) e nível socioeconômico.

**Variáveis Quantitativas**

Representam medidas numéricas ou quantidades observáveis que podem ser expressas em números. Elas são resultantes de contagens (conjunto finito de valores) ou mensurações (conjunto infinito de valores). Estas variáveis podem ser **discreta**s ou **contínuas**.

**Discreta**


Uma variável quantitativa discreta é aquela que apresenta valores que podem ser contados e geralmente inteiros. Exemplos: quantidade de produto vendido, número de cômodos em uma casa e idade.

**Contínua**


Uma variável quantitativa contínua é aquela que apresenta valores que podem assumir qualquer valor dentro de um intervalo específico e geralmente são expressos como números reais.

Exemplos: peso de um produto, altura de uma pessoa, tempo e temperatura de um experimento.

# **Aula 05 - Ajustando os Visuais**

```python
import pandas as pd
import seaborn as sns

#Importando a base de dados

tmdb = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/tmdb_5000_movies.csv")

#Filtrando os dados para colunas desejadas

dados = tmdb[["title", "original_language", "budget", "revenue"]]
linguas = ["fr","es","de"]

#Mantendo dados com valores de receita e orcamento validos e linguas desejadas (fr,es,de)

dados_validos = dados.query('revenue > 0 and budget > 0 and original_language == @linguas')

#Construindo o grafico de dispersao mais a reta de regressao linear para cada casp
sns.lmplot(data = dados_validos, x = 'budget', y = 'revenue', col = 'original_language',
           hue = 'original_language', ci = None, height = 4, scatter_kws = {"s": 30, "alpha": 0.5})
```

```python
import matplotlib.pyplot as plt
total_de_outros_filmes = tmdb.query("original_language != 'en'").original_language.value_counts()
plt.figure(figsize=(16,8))
sns.countplot(data=tmdb.query("original_language != 'en'"),
              x='original_language',order=total_de_outros_filmes.index,
              hue='original_language', hue_order= total_de_outros_filmes.index,
              palette='mako',
              stat="percent",
              )

plt.title('Distribuição de Linguas Original dos Filmes')
plt.show()
```

```python
tmdb.query("original_language != 'en'").original_language.value_counts(normalize=True)
```

**parâmetros no countplot():**

**order:** como ordenar os dados. Aqui vamos ordenar de acordo com o índice em total_de_outros_filmes_por_lingua;

**hue:** determina qual coluna devemos utilizar para colorir o gráfico. Aqui, vamos passar o mesmo valor do eixo x original_language.

**pallete:** determina a paleta de cores utilizada para colorir

**hue_order:** responsável por ordenar as cores, a mesma ordenação dos dados (order)

**plt.figure(figsize(x,y)) :** responsavel por ajustar o tamanho da figura


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tmdb = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/tmdb_5000_movies.csv")
filmes_outras_linguas = tmdb.query("original_language != 'en'").original_language.value_counts()
plt.figure(figsize=(14,8))
sns.countplot(data=tmdb.query("original_language != 'en'"),
              order = filmes_outras_linguas.index,
              palette = "mako",
              hue = "original_language",
              hue_order = filmes_outras_linguas.index,
              x = "original_language",
              stat = "percent")
plt.title("Distribuição de linguas originais nao inglesas de filmes")
plt.show()
```

```python
# prompt: separe as notas dos votos(vote_average) de 0 a 3, de 3 a 6, e de 6 a 10 e exiba em um grafico de barros comparando com a receita(revenue)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar os filmes por faixa de nota e calcular a média da receita
tmdb = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/tmdb_5000_movies.csv")
tmdb['vote_average_range'] = pd.cut(tmdb['vote_average'], bins=[0, 3, 6, 10], labels=['0-3', '3-6', '6-10'])
revenue_by_rating = tmdb.groupby('vote_average_range')['revenue'].mean()


# Criar o gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=revenue_by_rating.index, y=revenue_by_rating.values)
plt.title('Média da Receita por Faixa de Nota')
plt.xlabel('Faixa de Nota (vote_average)')
plt.ylabel('Média da Receita (revenue)')
plt.show()
```
